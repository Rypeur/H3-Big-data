```python
import pandas as pd 
import numpy as np 
from sklearn import linear_model 
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sb

pd.set_option('display.max_columns', None)
```

```python
df_price = pd.read_csv('price_availability.csv', sep=";")
df_listing = pd.read_csv('listings_final.csv',sep=";")
```

```python
df_price = df_price.groupby('listing_id').mean()
df_price
```

```python
df = pd.merge(df_price, df_listing, on=['listing_id'])
```

```python
df
```

```python
df.describe()
```

```python
X = np.array([df.person_capacity, df.bedrooms, df.bathrooms])
y = np.array(df.local_price)

X.shape
```

```python
y.shape
```

```python
X = X.reshape((999,3))

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
reg = LinearRegression().fit(X_train,y_train)
reg
```

```python
print("Coefficients beta_j :\n{}\nCoefficients INTERCEPT beta_0 :\n{}".format(reg.coef_,reg.intercept_))
```

```python
reg1 = reg.fit(X_train,y_train)
y_predict = reg.predict(X_test)
y_predict.size
```

```python
print("Mean squared error: {}\nR2 : {}".format(mean_squared_error(y_test,y_predict),r2_score(y_test,y_predict)))
```
